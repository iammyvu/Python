### Notice that the data has a time dimension (year). For this assignment, forget about time and treat all the observations as if they're from the same year. ### Now, take into account the time factor. Replicate your second answer but this time fill in the missing values by using a statistic that is calculated within the year of the observation. For example, if you want to fill a missing value for a variable with the mean of that variable, calculate the mean by using only the observations for that specific year.

```python
used2 = used.copy()

to_fill = ['ENROLL', 'TOTAL_REVENUE', 'FEDERAL_REVENUE', 'STATE_REVENUE', 'LOCAL_REVENUE', 
           'TOTAL_EXPENDITURE', 'INSTRUCTION_EXPENDITURE', 'SUPPORT_SERVICES_EXPENDITURE', 
           'OTHER_EXPENDITURE', 'CAPITAL_OUTLAY_EXPENDITURE', 'GRADES_PK_G', 'GRADES_KG_G', 
           'GRADES_4_G', 'GRADES_8_G', 'GRADES_12_G', 'GRADES_1_8_G', 'GRADES_9_12_G', 'GRADES_ALL_G']

years = used2['YEAR'].unique()

for column in to_fill:
    for year in years:
        used2.loc[used2['YEAR']==year, column].fillna(used2[used2['YEAR']==year][column].mean(),inplace=True)
```

##### Since the data is a time series data, we will fill in the missing values using `interpolation`.

```python
states = used['STATE'].unique()

to_fill = ['ENROLL', 'TOTAL_REVENUE', 'FEDERAL_REVENUE', 'STATE_REVENUE', 'LOCAL_REVENUE', 
           'TOTAL_EXPENDITURE', 'INSTRUCTION_EXPENDITURE', 'SUPPORT_SERVICES_EXPENDITURE', 
           'OTHER_EXPENDITURE', 'CAPITAL_OUTLAY_EXPENDITURE', 'GRADES_PK_G', 'GRADES_KG_G', 
           'GRADES_4_G', 'GRADES_8_G', 'GRADES_12_G', 'GRADES_1_8_G', 'GRADES_9_12_G', 'GRADES_ALL_G']

for state in states:
    used.loc[used['STATE']== state, to_fill] = used.loc[used['STATE'] == state, to_fill].interpolate()

# drop nulls after interpolation
used.dropna(inplace=True)
```

### Consider the two variables: TOTAL_REVENUE and TOTAL_EXPENDITURE. Do these variables have outlier values?

```python
# use boxplot to look for outlier
plt.figure(figsize=(15,5))

plt.subplot(121)
plt.boxplot(used['TOTAL_REVENUE'], whis=1.5)
plt.title('Box plot of Total Revenue')

plt.subplot(122)
plt.boxplot(used['TOTAL_EXPENDITURE'], whis=1.5)
plt.title('Box plot of Total Expenditure')

plt.show()
```

### If you detect outliers in the TOTAL_REVENUE and TOTAL_EXPENDITURE variables, apply the techniques you learned in this checkpoint to eliminate them and validate that there's no outlier values after you handled them.

We will use winsorization to eliminate outliers.

```python
from scipy.stats.mstats import winsorize

# create new variables
# apply 1 way winsorization
# we'll set the values of outliers to the value of 95% percentile
winsorized_revenue = winsorize(used['TOTAL_REVENUE'], (0, 0.05))

winsorized_expenditure = winsorize(used['TOTAL_EXPENDITURE'], (0, 0.05))

plt.figure(figsize=(15,5))

plt.subplot(121)
plt.boxplot(winsorized_revenue, whis=5)
plt.title('Box plot of Winsorized Total Revenue')

plt.subplot(122)
plt.boxplot(winsorized_expenditure, whis=5)
plt.title('Box plot of Winsorized Total Expenditure')

plt.show()
```

All outliers are now eliminated after winsorization.

### Create another variable by subtracting the original TOTAL_EXPENDITURE from TOTAL_REVENUE (before you eliminated the outliers). You can think of it as a kind of budget deficit in education. Do you find any outlier values in this new variable? If so, eliminate them using the technique you think most suitable.

```python
deficit = used['TOTAL_REVENUE'] - used['TOTAL_EXPENDITURE']

plt.boxplot(deficit, whis=5)
plt.title('Box plot of Deficit')
plt.show()
```

An example of education's budget deficit is due rising personnel cost and that cost cannot be resolved by shifting money within the school system's budget. 

The deficit variable has many outliers on both tails, so let's winsorize it:

```python
winsorized_deficit = winsorize(deficit, (0.05, 0.05))

plt.boxplot(winsorized_deficit, whis=5)
plt.title('Box plot of Winsorized Deficit')
plt.show()
```

Now all outliers are eliminated.


### Now create another variable by subtracting the TOTAL_EXPENDITURE from TOTAL_REVENUE. This time, use the outlier eliminated versions of TOTAL_EXPENDITURE from TOTAL_REVENUE. In this newly created variable, can you find any outliers? If so, eliminate them.

```python
deficit1 = winsorized_revenue - winsorized_expenditure

plt.boxplot(deficit1, whis=5)
plt.title('Box plot of New Deficit')
plt.show()
```

There are outliers so let's eliminate them using winsorization:

```python
winsorized_deficit1 = winsorize(deficit1, (0.05, 0.05))

plt.boxplot(winsorized_deficit1, whis=5)
plt.title('Box plot of Winsorized New Deficit')
plt.show()
```

We successfully eliminate the high outliers but not the low outliers. If we increase the lower bound of winsorization, the remaining few ones also would be eliminated but for consistency with the analysis above, we keep winsorizing the values using 5% as the theshold percentiles at both ends.

### Notice there are too many missing values for math and reading scores. Fill out the missing values using mean, median, and linear interpolation. Then compare the effects of these techniques on the distributions of the score variables.

```python
plt.figure(figsize=(20,20))

plt.subplot(441)
plt.hist(used.AVG_READING_4_SCORE.dropna())
plt.title('histogram of average reading 4 score')

plt.subplot(442)
plt.hist(used.AVG_MATH_4_SCORE.dropna())
plt.title('histogram of average math 4 score')

plt.subplot(443)
plt.hist(used.AVG_READING_8_SCORE.dropna())
plt.title('histogram of average reading 8 score')

plt.subplot(444)
plt.hist(used.AVG_MATH_8_SCORE.dropna())
plt.title('histogram of average math 8 score')

plt.subplot(445)
plt.hist(used.AVG_READING_4_SCORE.fillna(used.AVG_READING_4_SCORE.mean()))
plt.title('histogram of mean filled average reading 4 score')

plt.subplot(446)
plt.hist(used.AVG_MATH_4_SCORE.fillna(used.AVG_MATH_4_SCORE.mean()))
plt.title('histogram of mean filled average math 4 score')

plt.subplot(447)
plt.hist(used.AVG_READING_8_SCORE.fillna(used.AVG_READING_8_SCORE.mean()))
plt.title('histogram of mean filled average reading 8 score')

plt.subplot(448)
plt.hist(used.AVG_MATH_8_SCORE.fillna(used.AVG_MATH_8_SCORE.mean()))
plt.title('histogram of mean filled average math 8 score')
         
plt.subplot(449)
plt.hist(used.AVG_READING_4_SCORE.fillna(used.AVG_READING_4_SCORE.median()))
plt.title('histogram of median filled average reading 4 score')

plt.subplot(4,4,10)
plt.hist(used.AVG_MATH_4_SCORE.fillna(used.AVG_MATH_4_SCORE.median()))
plt.title('histogram of median filled average math 4 score')

plt.subplot(4,4,11)
plt.hist(used.AVG_READING_8_SCORE.fillna(used.AVG_READING_8_SCORE.median()))
plt.title('histogram of median filled average reading 8 score')

plt.subplot(4,4,12)
plt.hist(used.AVG_MATH_8_SCORE.fillna(used.AVG_MATH_8_SCORE.median()))
plt.title('histogram of median filled average math 8 score') 
         
plt.subplot(4,4,13)
plt.hist(used.AVG_READING_4_SCORE.interpolate())
plt.title('histogram of interpolate reading 4 score')

plt.subplot(4,4,14)
plt.hist(used.AVG_MATH_4_SCORE.interpolate())
plt.title('histogram of interpolate math 4 score')

plt.subplot(4,4,15)
plt.hist(used.AVG_READING_8_SCORE.interpolate())
plt.title('histogram of interpolate reading 8 score')

plt.subplot(4,4,16)
plt.hist(used.AVG_MATH_8_SCORE.interpolate())
plt.title('histogram of interpolate math 8 score')    
         
plt.show()
```

The technique for filling in the missing values substantially affects the distributions of the variables. Especially in our case, filling with mean or median has the strongest effect. This is probably due to the fact that we have a lot of missing values in our data.

The change in the distribution might have some serious effects on our model results. Hence, instead of filling in the missing values with mean or median, we may fill them in using interpolation or we may just discard the missing values.
